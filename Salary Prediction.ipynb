## Import Important Library

```python
import pandas as pd #will use it to read our csv file and for data cleaning
```

## Loading Data

```python
extracted_data = pd.read_csv('Salary.csv')
```

```python
extracted_data.head()  # by default we will see the first 5 rows
```

```python
extracted_data.isnull().sum() # checking if we have any missing values
```

## Working With Clean Data

```python
clean_data = extracted_data #ensured we have clean data so we can assign our extracted data as clean data
```

## Splitting Data

```python
from sklearn.model_selection import train_test_split #useful to split our data the way we want
```

```python
x = clean_data['YearsExperience'].values.reshape(-1 , 1)
y = clean_data['Salary'].values

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.20 , random_state = 42)
```

## Training & Testing Data

```python
from sklearn.linear_model import LinearRegression #we will use simple machine learning supervised algorithm
```

```python
linear_regression_model = LinearRegression() # fitting our splitted data
linear_regression_model.fit(x_train , y_train)
```

```python
y_train_pred = linear_regression_model.predict(x_train)
```

```python
y_test_pred = linear_regression_model.predict(x_test)
```

## Evaluating Model

```python
from sklearn.metrics import r2_score # will help us with checkng how good our model is
```

```python
r2_train = r2_score(y_train , y_train_pred)
```

```python
r2_test = r2_score(y_test , y_test_pred)
```

```python
print('Training R^2 Score :' ,r2_train)
```

```python
print('Testing R^2 Score:' , r2_test)
```

## Visualizing Our Model

### *Importing Visualizing Library*

```python
import matplotlib.pyplot as plt 
```

```python
plt.scatter(x_test , y_test , color = 'green' , label = 'Exact')
plt.plot(x_test , y_test_pred , color = 'red' , label = 'Predicted')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary ($)')
plt.title('Linear Regression Visualization')
plt.legend()
plt.show()
```
